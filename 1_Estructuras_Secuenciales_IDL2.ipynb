**Ejercicio 1: Determina la velocidad final de un auto si su velocidad inicial es 0 km/h  y su aceleración es 0.8 m/s2; y el tiempo transcurrido es de 30 segundos.**

```python
velocidad_inicial = 0
aceleracion = 0.8
tiempo = 30

velocidad_final = velocidad_inicial + aceleracion * tiempo
velocidad_kmh = velocidad_final * 3.6

print("Velocidad final en m/s:", velocidad_final)
print("Velocidad final en km/h:", velocidad_kmh)
```

**Ejercicio 2: Un estudiante realiza cuatro exámenes, los cuales tienen la misma ponderación. Representen el algoritmo correspondiente para obtener el Promedio de las calificaciones obtenidas.**

```python


# Entrada de datos
examen1 = float(input("Ingrese la nota del primer examen: "))
examen2 = float(input("Ingrese la nota del segundo examen: "))
examen3 = float(input("Ingrese la nota del tercer examen: "))
examen4 = float(input("Ingrese la nota del cuarto examen: "))

# Proceso
promedio = (examen1 + examen2 + examen3 + examen4) / 4

# Salida de datos
print("El promedio de las calificaciones es:", promedio)
```

**Ejercicio 3: Tres personas deciden invertir su dinero para fundar una empresa. Cada una de ellas invierte una cantidad distinta. Obtener el porcentaje que cada uno invierte con respecto a la cantidad total invertida.**

```python
# Entrada de datos
inversion1 = float(input("Ingrese la inversión de la primera persona: "))
inversion2 = float(input("Ingrese la inversión de la segunda persona: "))
inversion3 = float(input("Ingrese la inversión de la tercera persona: "))

# Proceso
total_invertido = inversion1 + inversion2 + inversion3

porcentaje1 = (inversion1 / total_invertido) * 100
porcentaje2 = (inversion2 / total_invertido) * 100
porcentaje3 = (inversion3 / total_invertido) * 100

# Salida de datos
print("El total invertido es: S/", total_invertido)
print("Porcentaje de la primera persona:", porcentaje1, "%")
print("Porcentaje de la segunda persona:", porcentaje2, "%")
print("Porcentaje de la tercera persona:", porcentaje3, "%")
```

**Ejercicio 4: Un vendedor recibe su sueldo base más 25% extra por comisión de sus ventas; el vendedor desea saber cuánto dinero recibirá por concepto de
comisiones por las tres ventas que realiza en el mes y el total que recibirá
en el mes tomando en cuenta sueldo base y comisiones.**

```python
# Entrada de datos
sueldo_base = float(input("Ingrese el sueldo base del vendedor: "))
venta1 = float(input("Ingrese el monto de la primera venta: "))
venta2 = float(input("Ingrese el monto de la segunda venta: "))
venta3 = float(input("Ingrese el monto de la tercera venta: "))

# Proceso
total_ventas = venta1 + venta2 + venta3
comision = total_ventas * 0.25
sueldo_total = sueldo_base + comision

# Salida de datos
print("El total de ventas es: S/", total_ventas)
print("La comisión por ventas es: S/", comision)
print("El sueldo total del mes es: S/", sueldo_total)
```

**Ejercicio 5: Suponga que un individuo desea invertir su capital en un banco y desea saber cuánto dinero ganará después de un año si el banco paga a razón
de 2% mensual.**

```python
# Entrada de datos
capital = float(input("Ingrese el capital a invertir: "))

# Proceso
tasa_mensual = 0.02
meses = 12

ganancia = capital * tasa_mensual * meses
monto_final = capital + ganancia

# Salida de datos
print("El capital invertido es: S/", capital)
print("La ganancia después de un año es: S/", ganancia)
print("El monto final después de un año es: S/", monto_final)
```

**Ejercico 6: Realice un programa que representen el algoritmo donde se ingrese la edad de un apersona para determinar aproximadamente cuántos
meses, semanas, días y horas ha vivido una persona.**

```python
# Entrada de datos
edad = int(input("Ingrese la edad de la persona en años: "))

# Proceso
meses = edad * 12
semanas = edad * 52
dias = edad * 365
horas = dias * 24

# Salida de datos
print("La persona ha vivido aproximadamente:")
print(meses, "meses")
print(semanas, "semanas")
print(dias, "días")
print(horas, "horas")
```

**Ejercicio 7: Un estacionamiento requiere determinar el cobro que debe aplicar a las personas que lo utilizan. Considere que el cobro es con base en las
horas que lo disponen y que las fracciones de hora se toman como
completas y realice un programa que permita determinar el cobro. El
pago por hora es de S/5.00 y el pago por los minutos restantes de S/0.10
por minuto.**

```python
# Entrada de datos
horas = int(input("Ingrese la cantidad de horas utilizadas: "))
minutos = int(input("Ingrese la cantidad de minutos adicionales: "))

# Proceso
pago_horas = horas * 5.00
pago_minutos = minutos * 0.10
total_pagar = pago_horas + pago_minutos

# Salida de datos
print("El pago por horas es: S/", pago_horas)
print("El pago por minutos es: S/", pago_minutos)
print("El total a pagar es: S/", total_pagar)
```

**Ejercicio 8:Una empresa constructora vende terrenos con la forma A de la figura. Obtener el área respectiva de un terreno de medidas ingresando el valor
de los 3 lados.**

```python
# Entrada de datos
lado1 = float(input("Ingrese el primer lado del terreno: "))
lado2 = float(input("Ingrese el segundo lado del terreno: "))
lado3 = float(input("Ingrese el tercer lado del terreno: "))

# Proceso
semiperimetro = (lado1 + lado2 + lado3) / 2

area = (semiperimetro *
        (semiperimetro - lado1) *
        (semiperimetro - lado2) *
        (semiperimetro - lado3)) ** 0.5

# Salida de datos
print("El semiperímetro del terreno es:", semiperimetro)
print("El área del terreno es:", area)
```

**Ejercicio 9:Cuanto pagaría un cliente por una compra con descuento y cuanto sería el cambio que recibiría en un pago en efectivo. Ingresar el monto base,
el porcentaje de descuento y el con cuánto dinero está pagando el
cliente.**

```python
# Entrada de datos
monto_base = float(input("Ingrese el monto base de la compra: "))
porcentaje_descuento = float(input("Ingrese el porcentaje de descuento: "))
pago_cliente = float(input("Ingrese el dinero entregado por el cliente: "))

# Proceso
descuento = monto_base * porcentaje_descuento / 100
monto_pagar = monto_base - descuento
cambio = pago_cliente - monto_pagar

# Salida de datos
print("El monto base de la compra es: S/", monto_base)
print("El descuento aplicado es: S/", descuento)
print("El monto a pagar es: S/", monto_pagar)
print("El cambio que debe recibir el cliente es: S/", cambio)
```

**Ejercicio 10: Se necesita saber la cantidad de dinero que un cajero automático debe proporcionar en diferentes denominaciones, teniendo en cuenta que el usuario debe ingresar la cantidad a retirar y el cajero tiene disponible las siguientes denominaciones de dinero: 200, 100, 50, 20, 10, 5, 1. Se debe
buscar entregar la menor cantidad de dinero por parte del cajero.**
