## Ejercicio 1 — Suma y promedio de pares e impares del 1 hasta N

```python
# Ejercicio 1: Suma y promedio de pares e impares del 1 hasta N
N = 20  # dato simulado

suma_pares = 0
suma_impares = 0
cont_pares = 0
cont_impares = 0

for i in range(1, N + 1):
    if i % 2 == 0:
        suma_pares += i
        cont_pares += 1
    else:
        suma_impares += i
        cont_impares += 1

promedio_pares = suma_pares / cont_pares
promedio_impares = suma_impares / cont_impares

print(f'N = {N}')
print(f'\n--- PARES ---')
print(f'Suma    : {suma_pares}')
print(f'Promedio: {promedio_pares:.2f}')
print(f'\n--- IMPARES ---')
print(f'Suma    : {suma_impares}')
print(f'Promedio: {promedio_impares:.2f}')
```

## Ejercicio 2 — Clasificar 10 números (par/impar/positivo/neutro/negativo)

```python
# Ejercicio 2: 10 números — pares, impares, positivos, neutros, negativos
numeros = [4, -3, 0, 7, 12, -5, 0, 9, -2, 6]  # datos simulados

pares = impares = positivos = neutros = negativos = 0

print('Números ingresados:', numeros)
print()

for n in numeros:
    if n % 2 == 0:
        pares += 1
    else:
        impares += 1

    if n > 0:
        positivos += 1
    elif n == 0:
        neutros += 1
    else:
        negativos += 1

print(f'Pares    : {pares}')
print(f'Impares  : {impares}')
print(f'Positivos: {positivos}')
print(f'Neutros  : {neutros}')
print(f'Negativos: {negativos}')
```

## Ejercicio 3 — Porcentajes respecto a un valor X

```python
# Ejercicio 3: Porcentajes de números respecto a X
numeros = [3, 7, 10, 2, 10, 15, 5, 10, 1, 8]  # datos simulados
X = 10
N = len(numeros)

menores = mayores = iguales = entre1yX = ninguna = 0

for n in numeros:
    if n < X:
        menores += 1
    elif n > X:
        mayores += 1
    elif n == X:
        iguales += 1

    if 1 <= n <= X:
        entre1yX += 1

    if n >= X and n != X and n <= 1:
        ninguna += 1

# Números que no cumplen ninguna condición anterior
ninguna = N - menores - mayores - iguales
ninguna = max(ninguna, 0)

print(f'Números: {numeros}')
print(f'X = {X}  |  N = {N}')
print(f'\nMenores que X   : {menores/N*100:.1f}%')
print(f'Mayores que X   : {mayores/N*100:.1f}%')
print(f'Iguales a X     : {iguales/N*100:.1f}%')
print(f'Entre 1 y X     : {entre1yX/N*100:.1f}%')
print(f'Ninguna cond.   : {ninguna/N*100:.1f}%')
```

## Ejercicio 4 — Máximo y mínimo simultáneo de una lista

```python
# Ejercicio 4: Máximo y mínimo simultáneo
numeros = [34, 12, 78, 5, 56, 23, 90, 45, 3, 67]  # datos simulados

maximo = numeros[0]
minimo = numeros[0]

for n in numeros:
    if n > maximo:
        maximo = n
    if n < minimo:
        minimo = n

print(f'Números: {numeros}')
print(f'\nMáximo: {maximo}')
print(f'Mínimo: {minimo}')
```

## Ejercicio 5 — Ahorro mensual (12 meses)

```python
# Ejercicio 5: Ahorro mensual acumulado en 12 meses
depositos = [150, 200, 180, 220, 170, 300, 250, 190, 210, 260, 230, 280]  # datos simulados
meses = ['Ene','Feb','Mar','Abr','May','Jun','Jul','Ago','Sep','Oct','Nov','Dic']

acumulado = 0
print(f'{"Mes":<6} {"Depósito":>10} {"Acumulado":>12}')
print('-' * 30)

for i in range(12):
    acumulado += depositos[i]
    print(f'{meses[i]:<6} S/ {depositos[i]:>8.2f} S/ {acumulado:>10.2f}')

print('-' * 30)
print(f'Ahorro final del año: S/ {acumulado:.2f}')
```

## Ejercicio 6 — Cronómetro (hh:mm:ss) referencial

```python
# Ejercicio 6: Cronómetro referencial hh:mm:ss
import time

duracion = 10  # segundos a cronometrar (simulado, cambia a lo que quieras)

print('Cronómetro iniciado...')
inicio = time.time()

for _ in range(duracion):
    elapsed = int(time.time() - inicio)
    hh = elapsed // 3600
    mm = (elapsed % 3600) // 60
    ss = elapsed % 60
    print(f'\r{hh:02d}:{mm:02d}:{ss:02d}', end='')
    time.sleep(1)

elapsed = int(time.time() - inicio)
hh = elapsed // 3600
mm = (elapsed % 3600) // 60
ss = elapsed % 60
print(f'\nTiempo final: {hh:02d}:{mm:02d}:{ss:02d}')
```

## Ejercicio 7 — Factorial de dos números y comparación de múltiplos

```python
# Ejercicio 7: Factorial de dos números y hasta qué múltiplo coinciden
def factorial_detalle(n):
    resultado = 1
    pasos = []
    for i in range(1, n + 1):
        resultado *= i
        pasos.append(str(i))
    return resultado, 'x'.join(pasos)

a = 7   # dato simulado
b = 10  # dato simulado

fact_a, detalle_a = factorial_detalle(a)
fact_b, detalle_b = factorial_detalle(b)

print(f'{a}! = {fact_a}  -->  {detalle_a}')
print(f'{b}! = {fact_b}  -->  {detalle_b}')

# Hasta qué valor coinciden (el menor de los dos)
coincide = min(a, b)
print(f'\n>> Son iguales hasta el múltiplo {coincide}')
```

## Ejercicio 8 — Fiesta: estadísticas de asistentes

```python
# Ejercicio 8: Estadísticas de asistentes a una fiesta
asistentes = [
    {'nombre': 'Carlos',   'sexo': 'M', 'edad': 25},
    {'nombre': 'Maria',    'sexo': 'F', 'edad': 30},
    {'nombre': 'Luis',     'sexo': 'M', 'edad': 17},
    {'nombre': 'Ana',      'sexo': 'F', 'edad': 22},
    {'nombre': 'Pedro',    'sexo': 'M', 'edad': 45},
    {'nombre': 'Lucia',    'sexo': 'F', 'edad': 19},
    {'nombre': 'Jorge',    'sexo': 'M', 'edad': 38},
    {'nombre': 'Sofia',    'sexo': 'F', 'edad': 28},
]  # datos simulados

total = len(asistentes)
hombres = [a for a in asistentes if a['sexo'] == 'M']
mujeres = [a for a in asistentes if a['sexo'] == 'F']

prom_hombres = sum(h['edad'] for h in hombres) / len(hombres)
prom_mujeres = sum(m['edad'] for m in mujeres) / len(mujeres)

menor_edad = min(asistentes, key=lambda x: x['edad'])
mayor_edad = max(asistentes, key=lambda x: x['edad'])

print(f'Total de asistentes : {total}')
print(f'Hombres             : {len(hombres)}')
print(f'Mujeres             : {len(mujeres)}')
print(f'Promedio edad H     : {prom_hombres:.1f} años')
print(f'Promedio edad M     : {prom_mujeres:.1f} años')
print(f'Menor edad          : {menor_edad["nombre"]} ({menor_edad["edad"]} años)')
print(f'Mayor edad          : {mayor_edad["nombre"]} ({mayor_edad["edad"]} años)')
```

## Ejercicio 9 — Producto: 100 x 98 x 96 x ... x 1

```python
# Ejercicio 9: Producto usando arrays: 100 x 98 x 96 x ... x 2 x 1
# Nota: la serie baja de 100 a 1, paso -2, luego termina con x1

array = list(range(100, 1, -2)) + [1]  # [100, 98, 96, ..., 2, 1]

producto = 1
for n in array:
    producto *= n

secuencia = ' x '.join(str(n) for n in array)
print(f'Secuencia: {secuencia}')
print(f'\nProducto total: {producto}')
```

## Ejercicio 10 — Sistema de notas para N estudiantes

```python
# Ejercicio 10: Sistema de notas — N estudiantes (mín 5, máx 20)
# Nota aprobatoria mínima: 13

notas = [18, 12, 15, 9, 13, 7, 20, 14, 11, 16, 8, 13]  # datos simulados (12 estudiantes)
N = len(notas)
APROBATORIA = 13

# Validar rango
notas = [n for n in notas if 0 <= n <= 20]

mayor_nota = max(notas)
menor_nota = min(notas)
cant_mayor = notas.count(mayor_nota)
cant_menor = notas.count(menor_nota)
promedio = sum(notas) / N
pares = sum(1 for n in notas if n % 2 == 0)
impares = N - pares
aprobados = sum(1 for n in notas if n >= APROBATORIA)
desaprobados = N - aprobados
porc_aprobados = aprobados / N * 100
porc_desaprobados = desaprobados / N * 100

print(f'Notas ingresadas: {notas}')
print(f'N = {N} estudiantes')
print(f'\na) Mayor nota              : {mayor_nota}')
print(f'b) Menor nota              : {menor_nota}')
print(f'c) Alumnos con mayor nota  : {cant_mayor}')
print(f'd) Alumnos con menor nota  : {cant_menor}')
print(f'e) Promedio del salón      : {promedio:.2f}')
print(f'f) Notas pares             : {pares}')
print(f'g) Notas impares           : {impares}')
print(f'h) Alumnos aprobados       : {aprobados}')
print(f'i) Alumnos desaprobados    : {desaprobados}')
print(f'j) % aprobados             : {porc_aprobados:.1f}%')
print(f'k) % desaprobados          : {porc_desaprobados:.1f}%')
```
